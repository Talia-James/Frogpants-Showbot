```python
def get_numeric(val):
    if type(val) == str:
        if val.isnumeric():
            return int(val)
        else:
            return 0
    elif type(val) == float:
        return int(val)
    else:
        return val
```

```python
def localize(x):
    try:
        y = pd.Timestamp.tz_localize(x,None)
        return y
    except AttributeError:
        y = x
        return y
```
